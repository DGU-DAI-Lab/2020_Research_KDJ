```python
import numpy as np
from matplotlib import pyplot as plt

def df_dx1(x1, x2, p):
    return x1 - p*(4*x1 + 3*x2 - 4)

def df_dx2(x1, x2, p):
    return x2 - p*(3*x1 + 4*x2 + 2)

p = 0.1
epoch = 500

x = [ (1.0,   0.9) ]
ans = (22/7, -20/7)

for i in range(epoch):
    _x = (x[-1][0], x[-1][1], p)
    x.append((df_dx1(*_x), df_dx2(*_x)))

x, ans = np.array(x), np.array(ans)

print('x_'+str(len(x)-1)+'\t', x[-1])
print('정답\t', ans)
print('delta\t', x[-1] - ans)

# Plotting

plt.scatter(x[:,0], x[:,1])
plt.plot(x[:,0], x[:,1])

plt.scatter(*x[0], color='black') # 초기값
plt.scatter(*ans, color='red')   # 정답

plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```
